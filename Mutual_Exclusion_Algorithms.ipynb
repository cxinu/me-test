```python
import pandas as pd
import numpy as np


df = pd.read_csv('output.csv')

avg_times_TAS = []
worst_times_TAS = []

avg_times_CAS = []
worst_times_CAS = []

avg_times_bounded = []
worst_times_bounded = []

# Loop over the DataFrame in chunks of 5 rows
for i in range(0, df.shape[0], 5):
    chunk = df.iloc[i:i+5]

    average = chunk.mean().tolist()

    # Append the average to the lists
    avg_times_TAS.append(average[0])
    worst_times_TAS.append(average[1])

    avg_times_CAS.append(average[2])
    worst_times_CAS.append(average[3])

    avg_times_bounded.append(average[4])
    worst_times_bounded.append(average[5])

df
```

```python
import matplotlib.pyplot as plt

threads = [10, 20, 30, 40, 50]
plt.figure(figsize=(10, 6))
plt.plot(threads, avg_times_TAS, marker='o', label='Average Time')
plt.plot(threads, worst_times_TAS, marker='o', label='Worst-case Time')
plt.xlabel('Number of Threads')
plt.ylabel('Time to Enter CS (ms)')
plt.legend()
plt.grid(True)
plt.title('Time to Enter CS vs Number of Threads for TAS')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(threads, avg_times_CAS, marker='o', label='Average Time')
plt.plot(threads, worst_times_CAS, marker='o', label='Worst-case Time')
plt.xlabel('Number of Threads')
plt.ylabel('Time to Enter CS (ms)')
plt.legend()
plt.grid(True)
plt.title('Time to Enter CS vs Number of Threads for CAS')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(threads, avg_times_bounded, marker='o', label='Average Time')
plt.plot(threads, worst_times_bounded, marker='o', label='Worst-case Time')
plt.xlabel('Number of Threads')
plt.ylabel('Time to Enter CS (ms)')
plt.legend()
plt.grid(True)
plt.title('Time to Enter CS vs Number of Threads for Bounded Waiting CAS')
plt.show()
```
